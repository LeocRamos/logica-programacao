# Condicional

```python
# If
if 5 > 10:
    print('Verdade')
```

```python
# If/else
num1 = 0
num2 = 0
print('Olá Caro usuário, digite um primeiro valor: ')
num1 = int(input())

print('Agora digite um segundo valor: ')
num2 = int(input())

if num1 > num2:
    print('O primeiro valor é maior que o segundo')
else:
    print('O segundo valor é maior que o primeiro')
```

```python
# If / Elif / Else
num1 = 0
num2 = 0
print('Olá Caro usuário, digite um primeiro valor: ')
num1 = int(input())

print('Agora digite um segundo valor: ')
num2 = int(input())

if num1 > num2:
    print('O primeiro valor é maior que o segundo')
elif num1 == num2:
    print('Os valores são iguais')
else:
    print('O segundo valor é maior que o primeiro')
```

# Condicional com operadores lógicos

```python
num1 = 0
num2 = 0
num3 = 0
print('Olá Caro usuário, digite um primeiro valor: ')
num1 = int(input())

print('Agora digite um segundo valor: ')
num2 = int(input())

print('Por fim, digite um terceiro valor: ')
num3 = int(input())
if (num1 > num2) and (num1 > num3):
    print('O primeiro valor é o maior')
else:
    print('O primeiro valor não é o maior')
```
